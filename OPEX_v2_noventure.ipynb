```python
def normalize(DATA,target_,target_list,driver_list):
  normalized_data = pd.DataFrame();
  for target in target_list:
    #drivers_2020_dict = []; 
    drivers_ = [];
    for driver in driver_list:
      driver_ = []; 
      data = DATA[DATA[target_]==target][driver].tolist()[0];
      acc = 0;
      for x in range(len(data)):
        if sum(data) == 0:
          acc = 0;
        else:
          acc =  (data[x] - min(data))/(max(data)-min(data))
        driver_.append(acc);
      drivers_.append(driver_);
    normalized_data = normalized_data.append({target_:target,
                            'Energía':drivers_[0],
                            'Producción':drivers_[1],
                            'Servicio a Pozo':drivers_[2],
                            'Workover':drivers_[3],
                            'Integridad':drivers_[4],
                            'Mantenimiento Civil':drivers_[5],
                            'Mantenimiento Civil Operaciones':drivers_[6],
                            'Mantenimiento Industrial':drivers_[7],
                            'Mantenimiento Ambiental':drivers_[8],
                            'Mantenimiento Ambiental Operaciones':drivers_[9],
                            'IT':drivers_[10],
                            'Ambiental':drivers_[11],
                            'Gestion de Tierras':drivers_[12],
                            'Gestión Social':drivers_[13],
                            'Proteccion Industrial':drivers_[14],
                            'Seguridad Industrial':drivers_[15],
                            'Otros HSSE':drivers_[16],
                            'Otros':drivers_[17],
                            'Total':drivers_[18]},ignore_index=True)
  return normalized_data;  
```

```python
def denormalize(NDATA,RDATA,target_,target_list,driver_list):
  denormalized_data = pd.DataFrame();
  for target in target_list:
    #drivers_2020_dict = []; 
    drivers_=[];
    for driver in driver_list:
      driver_ = []; 
      ndata = NDATA[NDATA[target_]==target][driver].tolist()[0];
      rdata = RDATA[RDATA[target_]==target][driver].tolist()[0]
      acc = 0;
      for x in range(len(ndata)):
        if sum(rdata) == 0:
          acc = 0;
        else:
          acc = ndata[x]*(max(rdata)-min(rdata)) + min(rdata);
        driver_.append(acc);
      drivers_.append(driver_);
    denormalized_data = denormalized_data.append({target_:target,
                            'Energía':drivers_[0],
                            'Producción':drivers_[1],
                            'Servicio a Pozo':drivers_[2],
                            'Workover':drivers_[3],
                            'Integridad':drivers_[4],
                            'Mantenimiento Civil':drivers_[5],
                            'Mantenimiento Civil Operaciones':drivers_[6],
                            'Mantenimiento Industrial':drivers_[7],
                            'Mantenimiento Ambiental':drivers_[8],
                            'Mantenimiento Ambiental Operaciones':drivers_[9],
                            'IT':drivers_[10],
                            'Ambiental':drivers_[11],
                            'Gestion de Tierras':drivers_[12],
                            'Gestión Social':drivers_[13],
                            'Proteccion Industrial':drivers_[14],
                            'Seguridad Industrial':drivers_[15],
                            'Otros HSSE':drivers_[16],
                            'Otros':drivers_[17],
                            'Total':drivers_[18]},ignore_index=True)
  return denormalized_data;
```

```python
def get_metrics(RDATA,PDATA,target_,target_list,driver_list):
  mean_rmse = 0;
  mean_rmse_data = pd.DataFrame();
  rmse_dataframe = pd.DataFrame();
  rmse_target = [];
  for target in target_list: 
    rmse_list = [];
    for driver in driver_list:
      if driver  in RDATA.columns and target in RDATA[target_].unique().tolist(): 
        yhat = PDATA[PDATA[target_]==target][driver].tolist()[0][:6];
        y = RDATA[RDATA[target_]==target][driver].tolist()[0][:6];
        discriminant_list = [];
        for x in range(len(yhat)):
          lineal_diff= yhat[x]-y[x];
          sqr_diff = lineal_diff*lineal_diff;
          discriminant = sqr_diff/len(yhat)
          discriminant_list.append(discriminant)
        discriminant_integral = sum(discriminant_list);
        rmse = math.sqrt(discriminant_integral); 
        rmse_list.append(rmse);
      else:
        rmse = 0; 
        rmse_list.append(rmse)        
    rmse_dataframe = rmse_dataframe.append({target_:target,
                            'Energía':rmse_list[0],
                            'Producción':rmse_list[1],
                            'Servicio a Pozo':rmse_list[2],
                            'Workover':rmse_list[3],
                            'Integridad':rmse_list[4],
                            'Mantenimiento Civil':rmse_list[5],
                            'Mantenimiento Civil Operaciones':rmse_list[6],
                            'Mantenimiento Industrial':rmse_list[7],
                            'Mantenimiento Ambiental':rmse_list[8],
                            'Mantenimiento Ambiental Operaciones':rmse_list[9],
                            'IT':rmse_list[10],
                            'Ambiental':rmse_list[11],
                            'Gestion de Tierras':rmse_list[12],
                            'Gestión Social':rmse_list[13],
                            'Proteccion Industrial':rmse_list[14],
                            'Seguridad Industrial':rmse_list[15],
                            'Otros HSSE':rmse_list[16],
                            'Otros':rmse_list[17],
                            'Total':rmse_list[18]},ignore_index=True)
    rmse_target.append(rmse_list);
    #mean_rmse = stat.mean([i for i in rmse_target if i!=0] or [0]);
    
  for target in target_list: 
    rmse_list = [];
    for driver in driver_list: 
      if driver  in RDATA.columns and target in RDATA[target_].unique().tolist():       
        rmse_list.append(rmse_dataframe[rmse_dataframe[target_]==target][driver].item());
    if len(rmse_list) == 0: 
      mean_rmse = 0;
    else:
      mean_rmse = sum(rmse_list)/len(rmse_list);
    mean_rmse_data = mean_rmse_data.append({target_:target,'M_RMSE':mean_rmse},ignore_index=True)

  return [rmse_dataframe,mean_rmse_data];  
```
